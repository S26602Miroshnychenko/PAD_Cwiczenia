```python

#orders_df.pop("tshirt_category")

def clean(text):
  text=text.replace("Wh ","White ")
  text=text.replace(" White","White")
  text=text.replace("Tshirt","T-Shirt")
  text=text.replace("Bl ","Black ")
  text=text.replace(" Shirt"," T-Shirt")
  return text
#orders_df = orders_df.to_frame()
orders_df.tshirt_category=orders_df.tshirt_category.apply(clean)
orders_df.head()
```

```python
orders_df.tshirt_category.unique()
```

```python
def clothfunc(text,myindex):
  myarr = text.split(' ')
  if len(myarr) == 3:
    tshirt_color=myarr[0]
    tshirt_type=myarr[1]
    tshirt_size = myarr[2]  
  if len(myarr) == 2:
    tshirt_type=text
    tshirt_size=np.NaN
    tshirt_color=np.NaN
  if len(myarr) == 1:
    tshirt_type=text
    tshirt_color=np.NaN
    tshirt_size=np.NaN
  if len(myarr) == 0:
    tshirt_type=np.NaN
    tshirt_color=np.NaN
    tshirt_size=np.NaN
  return pd.DataFrame([[tshirt_type,tshirt_size,tshirt_color]], index=myindex)


orders_df[['tshirt_type','tshirt_size','tshirt_color']]= orders_df.tshirt_category.apply(clothfunc, myindex=orders_df.index)
```
